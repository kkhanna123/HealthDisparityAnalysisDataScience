# Combining all the cleaned datasets

```python
import pandas as pd
import numpy as np
```

Load all the cleaned data

```python
healthDF = pd.read_csv('../Clean Data/healthcareData.csv')
socioEcDF = pd.read_csv('../Clean Data/socioeconomicData.csv')
educationDF = pd.read_csv('../Clean Data/educationData.csv')
```

Merge data together

```python
combinedHealthcareAndSocioEconomicDF = pd.merge(healthDF,socioEcDF,on='tractID')
```

```python
finalDF = pd.merge(combinedHealthcareAndSocioEconomicDF,educationDF,on='tractID')
```

Validate final dataframe and do a little bit of cleaning

```python
finalDF.head()
```

```python
finalDF.info()
```

Remove the null values because there don't seem to be that many

```python
finalDF = finalDF.dropna()
```

```python
finalDF.info()
```

Drop the duplicate columns \(Keep State to verify a successful merge\), and reset the index


```python
finalDF.drop('totalPop',axis=1)
finalDF.drop('county',axis=1)
finalDF.reset_index(drop=True)
```

Verify a successful merge by ensuring that the 'state' and 'stateName' columns hold the same values


```python
### If all elements are equal, the sum is eual to the length, so printing 1 means that the merge was successful
print(sum(finalDF['stateName']==finalDF['state'])/len(finalDF['stateName']))
```

Save the ready-to-use data as a new file



```python
finalDF.to_csv('../Clean Data/finalDataset.csv', index=False)
```
